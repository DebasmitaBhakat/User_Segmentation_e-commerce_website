**Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Importing Dataset**

```python
data_set = pd.read_csv('Amazon_data.csv')
```

# Considering Income(x axis) and rating(y axis)

```python
X = data_set.iloc[: , 3:5].values
```

**Optimal number of clusters via elbow method**

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
  kmeans = KMeans( n_clusters= i, init= 'k-means++', random_state= 21)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title("WCSS via elbow method")
plt.xlabel("number of clusters")
plt.ylabel("WCSS value")

```

**K Means Model training on training set**

```python
kmeans= KMeans(n_clusters= 4, init = 'k-means++', random_state= 42)
y_means= kmeans.fit_predict(X)
```

```python
print(y_means)
```

**Visualizing Clusters**

```python
plt.scatter( X[ y_means==0,0], X[  y_means==0, 1], s=25, c= 'magenta', label= 'cluster 1')
plt.scatter( X[ y_means==1,0], X[  y_means==1, 1], s=25, c= 'blue', label= 'cluster 2')
plt.scatter( X[ y_means==2,0], X[  y_means==2, 1], s=25, c= 'red', label= 'cluster 3')
plt.scatter( X[ y_means==3,0], X[  y_means==3, 1], s=25, c= 'green', label= 'cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s= 100, c= 'black', label='Centroids')
plt.title("Cluster of Amazon Users")
plt.xlabel("Annual Income in INR")
plt.ylabel("Purchase Rating")
plt.legend()
plt.show()
```
